```python
# Some summary statistics
pd.options.display.max_columns = None
df.describe()
# 'Children' column shows 119206 rows - with 4 null values, rest all columns show the row count as 119210 (as in df.shape)
# Mean value shows 37% cancellations
# Lead_Time - Min - 0 days(the booking and stay started the same day. Max - 737 days(almost 2 years ahead of stay))
# Mean & 50th % seems to be showing skewed distribution
# Avg_Daily_Rate - Max amount - 5400, Min - In Negative, whereas Mean & 50th % seems to be close enough to be called even distribution
```

```python
px.histogram(df, x='lead_time')

# Distribution of the lead_time data with respect to the Bookings - 
# Most of the Bookings and stay happened on the same day, with the lead_time showing 0 value,
# the Count of Bookings goes down with the increase in lead_time
```

```python
# How many bookings were cancelled
no_canceled_bookings = df['is_canceled'].sum()
pct_canceled_bookings = df['is_canceled'].mean()
print(f"{no_canceled_bookings} bookings were canceled, which is {pct_canceled_bookings*100:.2f} % of all bookings")
```

```python
# Checking the details of canceled bookings
Cancelations = df.filter(['arrival_date_month','is_canceled']).groupby(by='arrival_date_month',as_index=False).sum()

Cancelations
```

```python
#Create bar chart of cancellations per month
px.bar(Cancelations, x="arrival_date_month", y="is_canceled")

#Fewer Cancelations happening in winter months? But could be because of lower count of bookings itself?
```

```python
# Check number of cancelations for the number of Bookings
# Calculate and plot total bookings every arrival month
Total_Bookings = df.filter(['arrival_date_month','is_canceled']).groupby(by='arrival_date_month',as_index=False)\
.count().rename(columns={'is_canceled':'total_bookings'})
Total_Bookings
```

```python
px.bar(Total_Bookings, x='arrival_date_month', y='total_bookings')
```

```python
# Calculate cancelations rates every month
# The 2 grouped data of cancelations and bookings can be merged to calculate the cancelation ratings
merged = pd.merge(Cancelations,Total_Bookings, on='arrival_date_month')
merged['cancelation_rate'] = merged['is_canceled']/merged['total_bookings']
merged
```

```python
#Plotting for the cancelations rates per arrival month
px.bar(merged, x='arrival_date_month', y='cancelation_rate')
# Arrival month doesnt seems to be very strong predictor for the cancelation rates 
```

```python
# Checking if the number of nights per stay influence the cancelation rates
# Creating a new column to total the weekday and weekend nights columns

df_total_nights = df.assign(total_nights=df['stays_in_week_nights'] + df['stays_in_weekend_nights'])\
.query('total_nights < 15')# to keep the outliers out - counting max stay nights to be < 15
```

```python
# Plot by Histgram
px.histogram(df_total_nights, x='total_nights', color='is_canceled', barmode='group')
# Still not a big influencer on the cancelation rates
```

```python
# Calculate cancelations by total stays
total_bookings = df_total_nights.filter(['total_nights','is_canceled'])\
.groupby(by='total_nights',as_index=False)\
.count()\
.rename(columns={'is_canceled':'total_bookings'})
cancellations = df_total_nights.filter(['total_nights','is_canceled'])\
.groupby(by='total_nights', as_index=False)\
.sum()

merged = pd.merge(total_bookings, cancellations, on='total_nights')
merged['ratio_canceled'] = merged['is_canceled'] / merged['total_bookings']

px.bar(merged, x='total_nights', y='ratio_canceled')
# Not much corelation by this too
```

```python
#Box Plot to check the distribution by the total stays...
px.box(df_total_nights, y='total_nights', color = 'is_canceled')
# Distribution in this data too not revealing much
```

```python
# Checking the correlation by Avg_daily_rate column

df_avg_daily_rate = df.query('avg_daily_rate < 1000') # keeping the outliers out 
px.box(df_avg_daily_rate, y='avg_daily_rate', color='is_canceled')
```

```python
# Checking the correlation plotting to get leads..
df.corr()
```

```python
# First lead seems to be between lead_time and cancelations. We can do a Boxplot
px.box(df, y='lead_time', color='is_canceled')
```

```python
# This box plot seems to be showing distribution which can be interpreted as the higher the lead time
# meaning the earlier the booking is done before the actual stay there seems to be a tendency to cancel the booking
# May be earlier the booking is done there are more opportunities to cancel the booking & change of plans. 
# Lesser the lead time of the booking , the customers are forced/tend to keep the booking
```

```python
# Another lead seems to be Deposit type (Full, None or partial) with cancelations,
# lets compute a frequency table which can summarize the distribution of values in the data 
# and provides insights into the relationship.

pd.crosstab(df['is_canceled'], df['deposit_type'])
```

```python
# Some insights...
# More Customers who paid 'Full Deposit' seems to have canceled their booking? 
# You would expect them to keep the booking so to not loose the deposit

# More Customers who didnt have anything to loose because no deposit was made, seems to have kept their bookings
# compared to those who canceled them with no deposit made.
```
