```python
# The column headers needs to be renamed to be more meaningful names as below
# Create a dictionary object to hold the new column names & replace them in the dataframe

new_names =  {'Unnamed: 0': 'Country',
              '? Summer': 'Summer Olympics',
              'Gold': 'Gold(Summer)',
              'Silver': 'Silver(Summer)',
              'Bronze': 'Bronze(Summer)',
              '? Winter': 'Winter Olympics',
              'Gold.1': 'Gold(Winter)',
              'Silver.1': 'Silver(Winter)',
              'Bronze.1': 'Bronze(Winter)',
              '# Games': 'Total Games', 
              'Gold.2': 'Gold(Total)',
              'Silver.2': 'Silver(Total)',
              'Bronze.2': 'Bronze(Total)'}

df.rename(columns = new_names, inplace = True)
df.head(10)
```
