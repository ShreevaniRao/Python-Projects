```python
# Requests - It is a python library that makes it easy to send HTTP requests. 
# It is generally used in web scraping projects to fetch the HTML and XML code of a website. 
# It also works with a Beautiful Soup library.

# Import the necessary libraries

import requests
from bs4 import BeautifulSoup

# Documentation URL for Requests Library for Python...
URL = "https://requests.readthedocs.io/en/latest"

# Make a request to the website
r = requests.get(URL)

print(r.content)
```

```python
#The extracted content is unstructured, as you can see. Beautiful Soup library is used with a request to scrape any data from these unstructured forms 
# and make it meaningful by using data analysis tools.
# By using the power of data visualization, we will use html.parser to specify the HTML content that should be parsed. 
# Then we will assign our result to the Soup.
# This Soup object now contains the parsed HTML content of the page.

# Parse the HTML content of the page

soup = BeautifulSoup(requests.get(URL).content, 'html.parser')

# Prints the contents in a readable structure
print(soup.get_text())
```
