```python
# function to get video statsistics details using API call
def get_video_details(video_id):
    stat_url = "https://www.googleapis.com/youtube/v3/videos?key="+API_KEY+"&id="+video_id+"&part=statistics"
    stat_response = requests.get(stat_url).json()
        
    view_count = stat_response['items'][0]['statistics']['viewCount']
    like_count = stat_response['items'][0]['statistics']['likeCount']
    comment_count = stat_response['items'][0]['statistics']['commentCount']
    
    return view_count, like_count,comment_count
```

```python
# function to get list of videos using API call for a Channel
def get_videos():
    # Make API Call
    pageToken=""
    url = "https://www.googleapis.com/youtube/v3/search?key="+API_KEY+"&channelId="+CHANNEL_ID+"&part=snippet,id&order=date&maxResults=10000"
    response = requests.get(url).json()
    
    # delay till the response object is available
    time.sleep(1)

    # create an array of dictionaries
    videos_list = []
    for video in response['items']:
        if video['id']['kind'] == "youtube#video":
            video_id = video['id']['videoId']

            video_title =video['snippet']['title'] 
            video_title =str(video_title).replace("&amp;","")

            upload_date = video['snippet']['publishedAt']
            upload_date= str(upload_date).split("T")[0]

            #call the function to retrieve the video details
            view_count, like_count, comment_count = get_video_details(video_id)

            # append to the array
            videos_list.append({'video_id':video_id,'video_title':video_title,
                            'upload_date':upload_date,'view_count':view_count,
                            'like_count':like_count,'comment_count':comment_count})
    
    # create a dataframe with the array
    df = pd.DataFrame.from_records(videos_list)
    return df
```
